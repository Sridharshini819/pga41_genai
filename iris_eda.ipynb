# Iris Dataset Exploratory Data Analysis

This notebook performs an end-to-end exploratory analysis of the `data/Iris.csv` dataset, including data quality checks, summary statistics, species-level comparisons, and visual insights.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

iris_path = "data/Iris.csv"
iris = pd.read_csv(iris_path)
iris.head()
```

## Dataset Overview

The dataset contains measurements for Iris flowers with the following columns: `SepalLengthCm`, `SepalWidthCm`, `PetalLengthCm`, `PetalWidthCm`, and `Species`. The goal is to understand the structure, identify patterns by species, and surface any anomalies or relationships between features.

```python
print("Shape:", iris.shape)
print("Columns:", iris.columns.tolist())
print("Data types:")
print(iris.dtypes)
```

## Data Quality Checks

Check for missing values, duplicate rows, and any inconsistencies in the species labels.

```python
missing_values = iris.isnull().sum()
duplicate_rows = iris.duplicated().sum()
species_counts = iris['Species'].value_counts()

print("Missing values by column:")
print(missing_values)
print("Number of duplicate rows:", duplicate_rows)
print("Species counts:")
print(species_counts)
```

## Summary Statistics

View the overall numeric distribution and species-specific summary statistics for the measurement features.

```python
overall_description = iris.describe()
species_description = iris.groupby('Species').describe()

display(overall_description)
display(species_description)
```

## Species Distribution

Examine class balance and the number of samples for each Iris species.

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=iris, x='Species', palette='pastel')
plt.title('Iris Species Distribution')
plt.xlabel('Species')
plt.ylabel('Count')
plt.tight_layout()
```

## Pairwise Relationships

Use pairwise plots to visualize the relationships between features and highlight cluster separation by species.

```python
sns.pairplot(iris.drop(columns=['Id']), hue='Species', corner=True, palette='bright')
plt.suptitle('Pairwise Feature Relationships by Species', y=1.02)
```

## Feature Distributions by Species

Visualize distributions and variation in sepal and petal measurements across species.

```python
plt.figure(figsize=(14, 10))
for i, feature in enumerate(['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm'], 1):
    plt.subplot(2, 2, i)
    sns.boxplot(data=iris, x='Species', y=feature, palette='Set2')
    plt.title(f'{feature} Distribution by Species')
    plt.xlabel('Species')
    plt.ylabel(feature)
plt.tight_layout()
```

```python
plt.figure(figsize=(14, 10))
for i, feature in enumerate(['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm'], 1):
    plt.subplot(2, 2, i)
    sns.violinplot(data=iris, x='Species', y=feature, palette='Set3', inner='quartile')
    plt.title(f'{feature} Violin Plot by Species')
    plt.xlabel('Species')
    plt.ylabel(feature)
plt.tight_layout()
```

## Correlation Analysis

Analyze correlation strengths between numeric features and identify the most strongly related measurement pairs.

```python
corr_matrix = iris.drop(columns=['Id']).corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', square=True, linewidths=.5)
plt.title('Correlation Matrix for Iris Measurements')
plt.tight_layout()
display(corr_matrix)
```

## Conclusions

- The dataset has no missing values and appears clean, with a balanced species distribution.
- Petal length and petal width show strong positive correlation, while sepal and petal measurements are less strongly correlated.
- `Iris-setosa` is well-separated from the other species in pairwise feature space, especially on petal dimensions.
- `Iris-versicolor` and `Iris-virginica` overlap more but can still be differentiated using a combination of petal and sepal features.
- Boxplots and violin plots indicate that `Iris-virginica` generally has larger sepal and petal measurements than the other species.
